### Popcorn Hack 1:

Q: 
``` java
public class ClubMembers {
    private ArrayList<MemberInfo> memberList;
    public void addMembers(String[] names, int gradYear) { }
    public ArrayList<MemberInfo> removeMembers(int year) { }
}
```
Based on the MemberInfo constructor, how would you create a new member named "Alice" graduating in 2024 who is in good standing?

A: new MemberInfo ("Alice", 2024, true)

Feedback: Correct! The constructor signature is MemberInfo(String name, int gradYear, boolean hasGoodStanding). You must provide all three arguments in the exact order specified.

### Popcorn Hack 2:

Q:
```java
public void addMembers(String[] names, int gradYear)
```
In addMembers, you need to iterate through the String[] names array. Which loop is the most elegant and least prone to off-by-one errors for this task?

A: for (String n : names)

Feedback: Correct! The enhanced for-loop (for-each loop) is perfect here because we just need to read each name and don't need the index. It completely avoids ArrayIndexOutOfBoundsException.

### Popcorn Hack 3:

Q:
```java
public ArrayList<MemberInfo> removeMembers(int year)
```
When removing elements from an ArrayList, why is it dangerous to use a standard forward loop for (int i = 0; i < list.size(); i++)?

Consider this scenario: You are at index 1 and you remove the element.

[0] A   [1] B (Remove)   [2] C   [3] D

A: The elements shift left, so the next element (C) moves to index 1. The loop increments to index 2, skipping C entirely.

Feedback: Correct! When you remove an element, all subsequent elements shift one position to the left. If your loop counter i increments, you will skip the element that just shifted into the current index. Solution: Iterate backwards for (int i = list.size() - 1; i >= 0; i--) or use an Iterator.
