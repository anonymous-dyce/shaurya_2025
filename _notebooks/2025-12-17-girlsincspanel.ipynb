On Tuesday, December 16th, I had the opportunity to attend the Girls in CS's DNHS Alumni Panel. As I look toward my own future in STEM, hearing their perspectives provided me with three major takeaways.

### 1. The "User-First" Mindset and Prototyping
One of the most interesting discussions centered on how project ideas are born. Tanisha mentioned her experience in a Human-Computer Interaction class, where the goal was to design solutions for sustainable environments.

The biggest lesson here wasn't just about the code, but the Software Developer Mindset: putting the user first. Whether it’s through multiple rounds of prototyping or geolocating personal photos (an idea Michael mentioned keeping on his phone), successful projects solve real problems. Michael’s internship experience with LLMs also reinforced that interviewing customers and researching their responses is the only way to ensure a feature actually provides value.

### 2. Communication is a Technical Skill
I feel like the panelists made it very clear that communication is just as important as syntax. Professors and employers look for meaningful commit messages and code comments so collaborators can actually follow your logic.Also, Rachit explained that his classes require creating complex diagrams for database mapping and JWT tokens. These visuals help "bridge the gap" for teammates who might not understand the raw code.

### 3. High School is a Launchpad
It was reassuring to hear how much our current curriculum at Del Norte (like CSP and CSA) prepares us for the rigor of college. The panelists told that the freedom to design our own projects here through skills given my Mr. Mort gives us a great advantage during college interviews.

They also emphasized that the "soft skills" we are learning now, such as time management, organization, and critical thinking, are the real tools that help handle the stress of a college workload.

## Final Reflections & Advice
The panel concluded with some powerful advice: Stay curious and stay connected. The tech field moves fast, and the tools we use today might be replaced by AI/ML tomorrow. By doing our own research, building a network of contacts, and looking for project-based opportunities, we can stay ahead of the curve.

Most importantly, the alumni reminded us to "stay relaxed." While high school feels high-stakes now, the passion for learning and the ability to take initiative are what will ultimately define our success.

#### One thing I would like to implement: 
I am definitely going to start keeping a "list of ideas" on my phone like Michael suggested, as I often get ideas for my project when I'm not directly working on it, and I have to remember it until I start working on the project again. I could message it to my group on Slack or save it in my notes app to be used later.
